```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,ElementClickInterceptedException
import warnings
warnings.filterwarnings('ignore')
```

```python
#connecting to the webdriver
driver=webdriver.Chrome(r"C:\chromedriver\chromedriver.exe")
```

```python
amazon='https://www.amazon.in/'
```

```python
name=input()
links=[]
Rating=[]
Review=[]
```

```python
def Amazon(amazon):
    driver.get(amazon)
    search_path=driver.find_element_by_xpath("//div[@class='nav-search-field ']/input")
    search_path.send_keys(name)
    search_button=driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
    search_button.click()
    for i in range(1,4):
        get_link=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-medium']/div[2]/div[2]/div/div/h2/a")
        for i in get_link:
            links.append(i.get_attribute('href'))
        try:
            nxt_btn=driver.find_element_by_xpath("//li[@class='a-last']/a")
            nxt_btn.click()
        except NoSuchElementException:
            pass
    
    for i in links:
        driver.get(i)
        print("Review & Rating Scrapping ►►►►►►►►►►►►►►►►►►►►►►►►")
        try:
            click_rating=driver.find_element_by_xpath("//a[1][@id='acrCustomerReviewLink']")
            click_rating.click()
        except NoSuchElementException or ElementClickInterceptedException:
            print("No rating avilable on this product ")
            pass
        
        try:
            click_reviews=driver.find_element_by_xpath("//a[@class='a-link-emphasis a-text-bold']")
            click_reviews.click()
        except NoSuchElementException:
            pass
        
        Star_page=1
        End_page=200
        for page in range(Star_page,End_page+1):
            try:
                Reviews=driver.find_elements_by_xpath("//div[@class='a-row a-spacing-small review-data']/span/span")
                for R in Reviews:
                    Product_Review.append(R.text.replace('\n',''))
            except NoSuchElementException:
                Review.append("Nan")
            try:
                Rating=driver.find_elements_by_xpath("//div[@class='a-section celwidget']/div[2]/a[1]")
                for i in Rating:
                    rat=i.get_attribute('title')
                    
                    Rating.append(rat[:3])
            except NoSuchElementException:
                Rating.append("Nan")
            print("Product review and rating  of page {} scraped ".format(page+1))
            try:
                next_page=driver.find_element_by_xpath("//div[@id='cm_cr-pagination_bar']/ul/li[2]/a")
                if next_page.text=='Next→':
                    next_page.click()
                    time.sleep(4)
            except NoSuchElementException:
                pass
            
        
        
```
